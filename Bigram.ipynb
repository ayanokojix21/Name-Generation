## Bigram

A **bigram** is a character-level sequencing model that predicts the next character based on the current one. Specifically, it estimates the probability of the second character given the first character.

```python
bigram_list = {}
for name in names:
  name = [''] + list(name) + ['']
  for c1, c2 in zip(name, name[1:]):
    bigram = (c1, c2)
    bigram_list[bigram] = bigram_list.get(bigram, 0) + 1
```

```python
# removing these names from names list
names = [name for name in names if '-' not in name and '.' not in name and ' ' not in name]
len(names)
```

```python
# No name generally have '-' or '.' so we can remove them from character list
chars = [char for char in chars if char not in ['-', '.', ' ']]
chars
```

```python
chars.append('') # we are adding start and end token
```

```python
char2idx = {ch: idx for idx, ch in enumerate(chars)}
char2idx
```

```python
idx2char = {idx: ch for ch, idx in char2idx.items()}
idx2char
```

```python
bigram_array = torch.zeros(len(idx2char), len(char2idx), dtype=torch.int32)
```

```python
for name in names:
  name = [''] + list(name) + ['']
  for ch1, ch2 in zip(name, name[1:]):
    idx1 = char2idx[ch1]
    idx2 = char2idx[ch2]
    bigram_array[idx1, idx2] +=1
```

```python
prob_matrix = (bigram_array + 1).float()
```

```python
prob_matrix.shape
```

```python
prob_matrix.sum(1, keepdims=True)
```

```python
prob_matrix /= prob_matrix.sum(1, keepdims=True)
prob_matrix[0].sum()
```

```python
generated_names = []
for i in range(10):
  out = []
  index = 0
  while True:
    probs = prob_matrix[index]
    index = torch.multinomial(probs, num_samples=1, replacement=True).item()
    out.append(idx2char[index])
    if index == 0:
      break
  generated_names.append(''.join(out))
```

```python
generated_names
```

```python
def generate_names():
  char = ''
  name = ''
  while True:
    probs = prob_matrix[char2idx[char]]
    char = idx2char[torch.multinomial(probs, num_samples=1, replacement=True).item()]
    if char == '':
      break
    name += char
  return name
```

```python
generate_names()
```

```python
log_likelihood = 0.0
n = 0

for char in names:
# for char in ['nishchal']:
  char = [''] + list(char) + ['']
  for ch1, ch2 in zip(char, char[1:]):
    idx1 = char2idx[ch1]
    idx2 = char2idx[ch2]
    prob = prob_matrix[idx1, idx2]
    log_prob = torch.log(prob)
    log_likelihood += log_prob
    n += 1

print(f'{log_likelihood=}')
negative_log_likelihood = -log_likelihood
print(f'{negative_log_likelihood=}')
print(f'{negative_log_likelihood/n}')
```
